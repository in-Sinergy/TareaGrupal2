```python
##importamos y leemos el database
import pandas as pd
UNLink = "https://docs.google.com/spreadsheets/d/e/2PACX-1vSPVJIoDhupNyUpYD65FSHt6kMmF_KaxE2RD6CV1FOMP4vVw7WfKQMqpqo0ieD9vhMAmYP1NpBmw0od/pub?gid=658985149&single=true&output=csv"
HumanDvp = pd.read_csv(UNLink)
HumanDvp
```

# 1. Arreglo de la tabla de datos

# 2. Limpieza de datos

```python
#Verificamos qué tipos de datos hay en cada columna
HumanDvp.info()
```

```python
#todas las columnas excepto Country deberían contener datos numéricos
#Verificamos presencia de caracteres no numéricos
set(HumanDvp.GNI.str.extractall('(\W)').loc[:,0].to_list())-set('.')
```

```python
#eliminamos las comas
HumanDvp['GNI']=HumanDvp.GNI.str.replace(",",'')
HumanDvp
```

```python
set(HumanDvp.GNI.str.extractall('(\W)').loc[:,0].to_list())-set('.')
```

```python
#quitamos los espacios que pueda haber
objectCols=HumanDvp.select_dtypes(include=['object']).columns
objectCols
```

```python
HumanDvp=HumanDvp.loc[:,objectCols].apply(lambda col:col.str.strip())
HumanDvp
#con esto, podemos asumir limpio el database
```

# 3. Formateo del database



```python
#revisamos el tipo de dato presente en cada columna
HumanDvp.info()
#todas excepto Country deberían ser numéricas
```

```python
#seleccionamos las columnas a convertir
objCol = ['Country']
numCol = list(set(HumanDvp.columns.to_list())- set(objCol))
numCol
```

```python
#definimos una función para convertir los datos de obj a numéricos
def objToNum (xNum):
    import re
    xNum=pd.to_numeric(xNum)
    return xNum
```

```python
#aplicamos la función a todas las columnas
HumanDvp[numCol]= HumanDvp.loc[:, numCol].applymap(objToNum)
HumanDvp
```

```python
#verificamos que haya cambiado el tipo de data
HumanDvp.info()
```

```python
#finalmente, trabajamos con la columna de texto (Country)
#buscamos duplicados
HumanDvp[HumanDvp.duplicated(subset='Country')]
```

```python
#ahora revisamos caracteres diferentes de letras
HumanDvp[HumanDvp.Country.str.contains(r"[^\w\s]")]
```

```python
#buscamos acentos
HumanDvp[HumanDvp.Country.str.contains(r"\w*[\u00C0-\u01DA']\w*")]
```

```python
#convertimos los símbolos encontrados a su equivalente en ascii
from unidecode import unidecode
HumanDvp['Country']=HumanDvp.Country.apply(unidecode)
```

```python
#por practicidad, se colocarán todos los nombres de países en mayúsculas
HumanDvp['Country']=HumanDvp.Country.str.upper()
HumanDvp
```

```python
#finalmente, revisamos los paréntesis
HumanDvp[HumanDvp.Country.str.contains(r"[(][^)]*[)]")]
```

```python
#como solo se encuentran especificaciones, se puede prescindir de ellas
HumanDvp['Country']=HumanDvp.Country.str.replace(pat=r"\(.*\)",repl="", regex=True)
```
